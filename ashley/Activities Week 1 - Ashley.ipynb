# Types of Data  

Different types of data require different ways to represent them. The two broad categories of data are categorical and numerical.

Categorical data is data that can be placed into groups due to a quality about it. Here are some examples:
- Colors: items can be placed into different groups based on their color (red, blue, and yellow).
- Genre of a Novel: novels can be grouped by their genre (Action, Romance, Science Fiction)

*Note: Categorical data can be represented through numbers, but the numbers do not represent a quantity. Consider these numbers to be just names of groups. For example, you can have group 1 and group 2, but that does not mean group 1 will have fewer items than group 2.*


Numerical data is data that represents quantities through numbers. Here are some examples:
- Weight: weight is represented by a range of numbers and can be added together.
- Number of tv shows: can subtract the number of tv shows of one genre from another genre.

*Math can be done on two pieces of numerical data to create something meaningful. For example, we can add together 2 different weights to understand the total weight. A counter example would be to add group 1 and group 2 together to get 3. This is not something we can use to learn something new.*

# Graphs

## Bar Charts
Bar charts are a way to display the relationship between a categorical vaiable and a numerical variable. There is a bar for each category you want to represent, and the length of the bar represents the numerical variable paired with each category. The width of the bins should be the same, so that you are only comparing the lengths of the bars and not their lengths and widths. You do not want to compare both, because we are only looking at only one numerical variable. We do not need to be concerned with another number.

```python
favorite = Table().with_columns("Favorite Color", make_array("Blue", "Green", "Red", "Purple"),
                                "Number of People", make_array(10, 5, 4, 11))
favorite.sort(1).barh("Favorite Color", "Number of People")
```

In this example, we use the categorical variable, "Favorite Color", and compare it with the numerical variable "Number of People". We are able to compare how popular certain colors are in comparison to others.

## Histograms

Histograms are a way to contextualize a single set of numerical values. Histograms give you a way to understand how you data set is spread out. The spread is determined by the bins that represent the data. Bins are ranges that group close values of the data together. 

(Side note: The widths of bins do not have to be equal. They can be unequal, the histogram will still have the same area, so long as the height is readjusted to different sized bins)

For example, one of your bins can be including all numbers between 10 and 20, and your next bin can hold numbers in the range of 20 to 30. *Bins do not include the last number in their range.*

Histograms are special, because they will calculate the percent per unit relative to your variable. This means that you can calculate the percentage of your data that is in a certain range. This percentage is the area of your histogram.


#### Area of Histograms

Area of a histogram is equal to the height of a bar multiplied by the width of the bar.  

Area = Width * Height 

Logic of Units: The area is equivalent to the percentage of the graph that a bin or multiple bins represents. In terms of the formula, area is the unit of width mutlitplied by the percent per unit on the height axis. The units cancel out to leave only the percent.

Percent = (Unit) * (Percent / Unit)


```python
weights = Table().with_columns("Weight", make_array(100, 101, 200, 150, 139, 180, 90, 125, 114, 160, 280, 145))
weights.hist("Weight", unit="Pounds", bins=np.arange(85, 310, 20))
```

In this histogram, we see the range of our distribution of weights. We can calculate the percentage of weights in each bin.

## Line Plots
Line plots are used to contextualize a numerical data set in regards to the passage of time. Theses plots take a single numerical data set, and plot it across a time axis. The x-axis is chronological, and allows us to draw conclusions based on the patterns we see.

```python
over_time = Table().with_columns("Year", np.arange(1950, 2020, 10),
                                 "Number of College Students (millions)", make_array(4.5, 6.7, 8, 10, 15, 17 , 20))
over_time.plot("Year", "Number of College Students (millions)")
```

From the graph, we can see a steady increase over time. Meaning that as time passes, more and more people are entering colleges and universities.

## Scatter Plots
Scatter plots are graphs that compare the relationshp between two different numerical data sets. The relationship between the two data sets is also called the association between the two variables. 

The association can be positive or negative. If the association between the variables is positve, then the slope is positive. Meaning the two variables are both increasing or both decreasing. If the association between the variables is negative, then the slope is negative. Meaning that one variable is increasing, while the other is decreasing.

The logic behind this is that:
- +/+ = +
- -/- = +

Or
- +/- = -
- -/+ = -

```python
debt = Table().with_columns("Number of College Students (millions)", make_array(4.5, 6.7, 8, 10, 15, 17 , 20),
                             "Student Debt (thousands)", make_array(1, 3, 5, 8, 20, 25, 36))
debt.scatter("Number of College Students (millions)", "Student Debt (thousands)")
```

### Introduction to Slopes

Slope is a way to measure how fast the relationship between 2 variables is changing. It is a rate of change between two different variables. 

This rate of change is captured in the formula: change of y divided by the change of x (or more commonly, rise divided by run).

slope = rise / run

Another way to calculate the slope is to compare two different points of a graph. The formula that uses two points on a graph is:

slope = (y2 - y1) / (x2 - x1)
